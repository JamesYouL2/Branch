```python
import json
import pandas as pd
import requests
```

```python
url = "https://randomuser.me/api/?results=500"
```

```python
response = requests.get(url)
data = response.json()
```

```python
df=pd.json_normalize(data["results"])
```

```python
df.rename(columns=lambda x: x.replace(".", "_"), inplace=True)
```

```python
df.to_csv("data.csv")
```
